## Hospital readmissions data analysis and recommendations for reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

B. I will setup hypothesis test where the null hypothesis will be: the mean of excess rate of readmission in the lower half of the discharge rate is equal to the number of excess rate of readmission in the upper half of the discharge rate. 

```python
#http://stackoverflow.com/questions/33367142/split-dataframe-into-relatively-even-chunks-according-to-length
# I am splitting the data into two the first half has less discharge and the second half has more discharge:
for g, df in clean_hospital_read_df.groupby(np.arange(len(clean_hospital_read_df["Number of Discharges"])) // 5789):
    print(df.shape)
```

```python
first = clean_hospital_read_df.iloc[:5789] 
second = clean_hospital_read_df.iloc[5789:]
```

```python
second.shape
```

```python
firstx = [a for a in first['Number of Discharges'][:]]
secondx = [a for a in second['Number of Discharges'][:]]
```

```python
plt.hist(firstx)
plt.title("Number of Discharge in the first half of the dataset")
plt.xlabel("Number of Discharge")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.hist(secondx)
plt.title("Number of Discharge in the second half of the dataset")
plt.xlabel("Number of Discharge")
plt.ylabel("Frequency")
plt.show()
```

```python
def TestStatistic(data):
    group1, group2 = data
    test_stat = abs(group1.mean() - group2.mean())
    return test_stat
```

```python
actual = TestStatistic((first, second))
actual
```

H_o: mean(excess admission ratio first) - mean(excess admission ratio second) = 0

H_1: mean(excess admission ratio first) - mean(excess admission ratio second) != 0

For $\alpha$ = .01, Null Hypothesis will be: mean(excess admission ratio first) - mean( excess admission ratio second) < 0.01. 

```python
sigma_P1_P2= np.sqrt(2*clean_hospital_read_df["Excess Readmission Ratio"].mean()/len(clean_hospital_read_df["Excess Readmission Ratio"]))
sigma_P1_P2
```

```python
z = (first["Excess Readmission Ratio"].mean() - second["Excess Readmission Ratio"].mean()) / sigma_P1_P2
print('z-score: {}'.format(z))
```

In order to reject the null hypothesis, the z-score has to be higher than 2.58. Our z-score is 1.06. Therefore, we cannot reject the null hypothesis.

__Statistical significance__: we cannot reject the null hypothesis. 

__Practical significance__: the mean excess rate of readmission is the same between the lower half of the population and the upper half of the population of discharge rate. So, we cannot use the number of discharges as the predictor for the excess readmission ratio.
